```python
# Import Dependencies

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load data files
# Load stat1

nba_data_stat1 = pd.read_csv('nba_team_data.csv')
nba_data_stat1.head()
```

```python
# Load stat2

nba_data_stat2 = pd.read_csv('nba_data.csv')
nba_data_stat2.head()
```

```python
# Clean-up stat1
# Remove NaN rows
# Change the Season from 20xx-yy to 20xx-20yy

# Remove NaN rows

nba_data_stat1_clean = nba_data_stat1.dropna(axis=0, how="any", subset=['Playoffs'])
nba_data_stat1_clean = nba_data_stat1_clean.reset_index(drop=True)
nba_data_stat1_clean.head()
nba_data_stat1_clean.count()
```

```python
# Change the Season from 20xx-yy to 20xx-20yy by adding "20" after the dash

season = nba_data_stat1_clean['Season']
i=0
rev_season = []

for row in season:
    season_word = season[i]
    rev_word = season_word.replace('-', '-20')
    rev_season.append(rev_word)
    i = i + 1
    
rev_season

nba_data_stat1_clean['Season'] = rev_season

nba_data_stat1_clean.head()
```

```python
# Merge the Season and Team Names to create unique identifier

nba_data_stat1_clean['Team ID'] = nba_data_stat1_clean['Team'] + nba_data_stat1_clean['Season']
nba_data_stat1_clean.head()

```

```python
# Remove Unnecessary columns for stat1 and name nba_stat1

nba_stat1 = nba_data_stat1_clean[['Team ID', 'W', 'L', 'W/L%','Finish','SRS','Pace','Rel_Pace','ORtg','Rel_ORtg','DRtg',
                                 'Rel_DRtg']]
nba_stat1.head()
nba_stat1.count()
```

```python
# Clean-up stat2

nba_data_stat2.head()
```

```python
nba_data_stat2['Team ID'] = nba_data_stat2['Team'] + nba_data_stat2['Year']
nba_data_stat2.head()
nba_data_stat2.count()
```

```python
# Capture necessary data and save as stat2
# Remove 2007-2008 rows

nba_data_temp = nba_data_stat2
nba_data_temp.count()

# nba_2018_teams = nba_data_stat2[nba_data_stat2.Year == '2017-2018']
# nba_2018_teams.head()
# nba_2018_teams = nba_2018_teams[['Team ID','Rk','MP','FG','FGA','FG%','3P','3PA','3P%','2P','2P%',
#                                 'FT','FTA','FT%','ORB','DRB','TRB','AST','STL','TOV','PF','PTS']]

# nba_2018_teams.head()
# nba_2018_teams = nba_2018_teams.reset_index(drop=True)
# nba_2018_teams.head()
# list(nba_2018_teams.columns.values)
```

```python
# Remove W/L columns

nba_data_temp2 = nba_data_temp.drop(['W','L'], axis=1)
nba_data_temp2.head()
nba_stat2 = nba_data_temp[['Team ID','Rk','MP','FG','FGA','FG%','3P','3PA','3P%','2P','2P%',
                                 'FT','FTA','FT%','ORB','DRB','TRB','AST','STL','TOV','PF','PTS','Year','Finals','Conference Finals',
                          '1st Round', '2nd Round']]
nba_stat2.head()
```

```python
# Merge 2 stat data

nba_stat_df = nba_stat1.merge(nba_stat2,how='outer', on='Team ID')
nba_stat_df.head()
# nba_stat_df.count()
```

```python
# 2018 PlayOff Teams

nba_2018_teams = nba_stat_df[nba_stat_df.Year == '2017-2018']
nba_2018_teams = nba_2018_teams.reset_index(drop=True)
nba_2018_teams.head()
```

```python
# Clean-up 2018 teams

nba_2018_teams = nba_2018_teams.drop(['Year','Finals','Conference Finals','1st Round','2nd Round'], axis=1)
nba_2018_teams.head()
nba_2018_teams.to_csv("nba_2018_teams.csv", sep=",", index=False)
```

```python
# First Round Winner Data
# nba first round teams

nba_playoff_df = nba_stat_df[nba_stat_df.Year != '2017-2018']
nba_playoff_df = nba_playoff_df.reset_index(drop=True)
nba_playoff_df.head()
```

```python
# Obtain Round1 Data

nba_playoff_rd1 = nba_playoff_df.drop(['Finals','Conference Finals', '2nd Round', 'Year'], axis=1)
nba_playoff_rd1.head()
# nba_playoff_rd1.count()

# This is our training data for 1st Round
nba_playoff_rd1.to_csv("nba_playoff_rd1.csv", sep=",", index=False)
```

```python
# Obtain Round2 Data

nba_playoff_rd2 = nba_playoff_df.drop(['Finals','Conference Finals', '1st Round', 'Year'], axis=1)
nba_playoff_rd2.head()
nba_playoff_rd2.to_csv("nba_playoff_rd2.csv", sep=",", index=False)
```

```python
# Obtain Conf Finals Data

nba_playoff_conf = nba_playoff_df.drop(['Finals','1st Round', '2nd Round', 'Year'], axis=1)
nba_playoff_conf.head()
nba_playoff_conf.to_csv("nba_playoff_conf.csv", sep=",", index=False)
```

```python
# Obtain Finals Data

nba_playoff_finals = nba_playoff_df.drop(['1st Round','Conference Finals', '2nd Round', 'Year'], axis=1)
nba_playoff_finals.head()
nba_playoff_finals.to_csv("nba_playoff_finals.csv", sep=",", index=False)
```

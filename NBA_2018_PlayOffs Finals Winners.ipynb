```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nba_data_temp = pd.read_csv('nba_playoff_finals.csv')
nba_data_temp = nba_data_temp.dropna(axis=0, how='any')
nba_data = nba_data_temp.drop(['Team ID'], axis=1)

nba_data.head()
```

```python
# Assign X (data) and z (target)

### BEGIN SOLUTION
X = nba_data.drop("Finals", axis=1)
z = nba_data["Finals"]
print(X.shape, z.shape)
### END SOLUTION
```

```python
### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, z_train, z_test = train_test_split(X, z, random_state=1, stratify=z)
### END SOLUTION
```

```python
### BEGIN SOLUTION
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
### END SOLUTION
```

```python
### BEGIN SOLUTION
classifier.fit(X_train, z_train)
### END SOLUTION
```

```python
### BEGIN SOLUTION
print(f"Training Data Score: {classifier.score(X_train, z_train)}")
print(f"Testing Data Score: {classifier.score(X_test, z_test)}")
### END SOLUTION
```

```python
### BEGIN SOLUTION
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {z_test[:10].tolist()}")
### END SOLUTION
```
